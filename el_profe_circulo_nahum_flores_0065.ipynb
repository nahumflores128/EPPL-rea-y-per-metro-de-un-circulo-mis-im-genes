```python
# calcular area y perimetro de un circulo
# variables A,P, r
# formulas A = pi*r^ y P = 2 * pi * r
# capturar el valor r
import math
print("Calcular el area y perimetro de un circulo")
r = float(input("Ingrese el valor de radio: "))
A = math.pi*r**2
P = 2 * math.pi*r
print("El area del circulo es: ", A)
print("El perimetro del circulo es: ", P)
print("Nahum flores NL = 18")
```
